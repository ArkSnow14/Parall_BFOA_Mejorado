```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('reporte_bacterias.xlsx')

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Fitness'], label='Fitness', marker='o', color='blue')
plt.plot(df.index, df['Blossum Score'], label='Blossum Score', marker='x', color='red')
plt.xlabel('Evaluación')
plt.ylabel('Valor')
plt.title('Evolución del Fitness y Blossum Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Best'].value_counts().sort_index().plot(kind='bar', color='green')
plt.xlabel('Índice Best')
plt.ylabel('Frecuencia')
plt.title('Distribución de los Índices Best Seleccionados')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['Seconds'], label='Tiempo (s)', marker='s', color='purple')
plt.xlabel('Evaluación')
plt.ylabel('Segundos')
plt.title('Tiempo de Ejecución por Evaluación')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(df.index, df['NFE'], label='NFE Acumuladas', marker='o', color='orange')
plt.xlabel('Evaluación')
plt.ylabel('NFE')
plt.title('Evaluaciones de Función Acumuladas')
plt.grid(True)
plt.show()
```
