```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('reporte_bacterias_SPS_ED_v4_recalc.csv')

plt.figure(figsize=(12, 6))
plt.plot(df['Iteration'], df['Fitness_Iter'], label='Fitness', marker='o')
plt.plot(df['Iteration'], df['SPS_Afin_Score_Iter'], label='SPS Afinidad', marker='x')
plt.xlabel('Iteración')
plt.ylabel('Valor')
plt.title('Evolución del Fitness y Puntaje SPS de Afinidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Iteration'], df['Interaction_Iter'], label='Interacción (Diferencia)', color='green', marker='s')
plt.xlabel('Iteración')
plt.ylabel('Diferencia')
plt.title('Evolución de la Interacción (Fitness - SPS Score)')
plt.axvline(x=10, color='r', linestyle='--', label='Cambio de fase (Iteración 11)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df['Iteration'], df['TimeSeconds_Iter'], color='purple')
plt.xlabel('Iteración')
plt.ylabel('Tiempo (segundos)')
plt.title('Tiempo de Ejecución por Iteración')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Iteration'], df['CumulativeNFE'], label='NFE Acumuladas', color='orange', marker='o')
plt.xlabel('Iteración')
plt.ylabel('NFE Acumuladas')
plt.title('Evaluaciones de Función Acumuladas (NFE)')
plt.grid(True)
plt.show()
```
